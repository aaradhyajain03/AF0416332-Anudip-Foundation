```python
#basic chart

import matplotlib.pyplot as plt

# Line Chart
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 40, 60]
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='*', linestyle='-', color='b', label='Line Chart')
plt.title('Line Chart Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 8, 5, 3]
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='g', label='Bar Chart')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Exam scores data
exam_scores = [22, 45, 60, 75, 80]

# Creating the histogram
plt.figure(figsize=(10,6))
plt.hist(exam_scores, bins=6, color='skyblue', edgecolor='black')
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
```
